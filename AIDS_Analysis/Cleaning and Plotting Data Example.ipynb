```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Below is the data obtained from UNAIDS about AIDS Related Deaths
```

```python
#Imported the data
death_data = pandas.read_csv("Data\AIDS-related deaths_Number of AIDS-related deaths_Population_ All ages.csv")
```

```python
death_data # Time to see how the data looks like
```

```python
# Looking at the column names, they contain whitespaces.
death_data.columns
```

```python
# Let's fix the column names since they contain whitespaces
death_data = death_data.rename(str.strip, axis = 'columns')
```

```python
# The column names now
death_data.columns
```

```python
# Now, cleaning up the useless data
for year in range(1990, 2017):
    death_data = death_data.drop(columns = [str(year) + '_Footnote', str(year) + '_lower',str(year) + '_upper'])
```

```python
death_data
```

```python
# Now, I will drop the countries with strings like <100, <500 etc 
for row_ in range(0, 169):
    for year in range(1990, 2017):
        if '<' in death_data[str(year)][row_]:
            death_data = death_data.drop(row_)
            break
```

```python
# Our data looks so much better now! 
death_data
```

```python
# Resetting the index since it gets messed up
death_data = death_data.reset_index(drop = True)
```

```python
# There's whitespaces in the values! Why!? Plus, there's tons of missing values, need to fix that.
for row_ in range(0, len(death_data.index)):
    for year in range(1990, 2017):
        death_data[str(year)][row_] = death_data[str(year)][row_].replace(' ', '')
        if(str(death_data[str(year)][row_]) == '...'):
            death_data = death_data.drop(row_)
            break
```

```python
death_data # So, we've fixed the entire data. Not yet converted to integer though, time to do that.
```

```python
death_data = death_data.reset_index(drop = True) # Fixing tyhe indexes since they've messed up.
```

```python
# Now, the entire data is converted into integer datatype.
for row_ in range(0, len(death_data.index)):
    for year in range(1990, 2017):
        death_data[str(year)][row_] = int ( death_data[str(year)][row_] )
```

```python
death_data # Let's look at the data! 
```

```python
labels = death_data['Country'].tolist()
death_data = death_data.drop(columns = 'Country')
```

```python
death_data
```

```python
# Okay, time to plot! 
plot_figure = plt.subplot()
for row_ in range(0, len(death_data.index) - 1):
    plot_figure.plot(death_data.columns.tolist(), death_data[row_:row_ + 1].values.tolist()[0], label = labels[row_])
```

```python
plot_figure.legend()
plt.show()
```
